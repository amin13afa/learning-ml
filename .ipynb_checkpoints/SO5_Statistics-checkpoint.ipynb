```python
ram_count = sp['Ram'].value_counts()
ram_cat = ram_count.index
ram_cat
```

```python
plt.figure(figsize=(6,4), dpi= 100)

plt.bar(ram_cat, ram_count, color = 'g')
plt.title("Ram\' distribution")
plt.xlabel("Ram")
plt.ylabel("faravani")
plt.xticks(ram_count.index)
plt.yticks(ram_count.values)
plt.show()
```

```python
os_count = sp['OS'].value_counts()
os_cat = os_count.index
```

```python
plt.figure(figsize= (6,4), dpi= 100)
plt.bar(os_cat, os_count, color = 'r')
plt.title("OS\'distribution")
plt.xlabel("OS")
plt.ylabel("faravani")
plt.xticks(os_count.index)
plt.yticks(os_count.values)
plt.show()
```

# ECDF 

Empirical cumulative distribution function

```python
def ECDF(data):
    n = len(data)             # Number of data point
    x = np.sort(data)         # x-data for ECDF
    y = np.arange(1,n+1)/n    # y-data for ECDF 
    return x, y
```

```python
x,y = ECDF(sp['inch'])
```

```python
plt.figure(figsize= (8,6), dpi= 100)
plt.scatter(x, y, color = 'g', s= 80)
plt.title("ECDF\ninch")
plt.xlabel("inch")
plt.ylabel("ECDF")
plt.xticks(sp['inch'])
plt.grid()
plt.margins()
plt.show()
```
