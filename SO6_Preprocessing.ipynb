## Concatenating

```python
dict1 = {'col1':['babak', 'raha', 'sara', 'reza'],
        'col2':[1,2,3,4],
        'col3':[13,14,17,12],
        'col4':[17,14,12,18],
        'col5':[17,15,20,19]}
my_source_one = pd.DataFrame(data= dict1)
my_source_one
```

```python
dict2 = {'col1': ['babak', 'baran', 'sara', 'arash', 'mahan','reza'],
        'col2' : [1,2,3,4,5,6],
        'col3' : [13, 15,14, 20, 15, 14],
        'col4' : [17, 20, 19, 19, 18, 12]}
my_source_two = pd.DataFrame(data= dict2)
my_source_two
```

```python
my_source_connat = pd.concat([my_source_one, my_source_two], axis= 0, ignore_index= True)
my_source_connat
```

```python
my_source_connat2 = pd.concat([my_source_one, my_source_two], axis= 1, ignore_index= True)
my_source_connat2
```

```python
my_source_connat.drop('col5', axis= 1, inplace= True)
```

```python
my_source_connat.drop_duplicates(inplace= True)
```

```python
my_source_connat.reset_index(drop= True, inplace= True)
```
