# Visualization

```python
# Import libraries
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Data 
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

# Line Plot 

```python
plt.plot(years, iran_pop, 'r--')
plt.show()
```

# Scatter Plot 

```python
plt.scatter(years, iran_pop)
plt.show()
```

# Histogram Plot 

```python
city_name = ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
city_pop = [7153309, 2307177, 1547164, 1448075, 1424641, 1249942]
```

```python
plt.hist(city_pop, bins= 'auto')
plt.show()
```

```python
# The "square root rule" is a commonly-used rule of thumb for choosing number of bins
# floor(np.sqrt(N))  where N: The number of data
plt.hist(city_pop, bins= 2)
plt.show()
```

```python
plt.hist(city_pop, bins= 'auto')
plt.show()
```

# Pie chart

```python
plt.pie(city_pop, labels= city_name)
plt.show()
```

# Customization plots 

Customize Line Plot

```python
plt.figure(figsize= (8,4), dpi= 100)  # dpi used for resolution 
plt.plot(years, iran_pop, color = 'g')
plt.title('Iran\'s Population')
plt.xlabel("Years")
plt.ylabel("Population")
plt.xticks(years)
plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],
           ['21M', '28M', '38M', '56M', '66M', '74M', '80M'])
plt.show()
```

```python
plt.figure(figsize=(10,8), dpi= 100)
plt.plot(years, iran_pop, 'g')
plt.title("Iran's \n Population")
# X customization
plt.xlabel("Years")
plt.xticks(years)
# Y customization
plt.ylabel("Population")
plt.yticks(iran_pop, ['21M', '28M', '38M', '56M', '66M', '74M', '80M'])

plt.show()
```

```python
ytic = [str(x)+'M' for x in iran_pop]
print(ytic)
```

# Customize Scatter plot

```python
plt.figure(figsize= (8,6), dpi= 100)
pop_size = np.array(city_pop)/10000
clr = ['violet', 'green', 'orange', 'tomato', 'blue', 'pink']

plt.scatter(np.arange(6), city_pop, s= pop_size, c= clr)
plt.title("City\'s Population'")
plt.xlabel("City")
plt.ylabel("Population")
plt.xticks([0, 1, 2, 3, 4, 5],
           ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz'])
plt.yticks([1000000, 2000000, 3000000, 7000000], 
          ['1M', '2M', '3M','7M'])
plt.margins(0.1)
plt.text(0, 7153309, 'Iran Capital', fontsize = 15 )
plt.show()
```

```python
plt.figure(figsize= (12,8), dpi= 100)
pop_size = np.array([7153309, 2307177, 1547164, 1448075, 1424641, 1249942])/1000
Colors = ['violet', 'green', 'orange', 'tomato', 'blue', 'pink']
plt.scatter(np.arange(6), city_pop, s= pop_size, c= Colors)
plt.title("Iran\'s Population")
plt.xticks(np.arange(6),
           ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz'])
plt.yticks([1000000, 2000000, 7000000], ['1M', '2M', '7M'])
plt.margins(0.15)
plt.text( 0,7153309, "Iran\'s Capital", fontsize = 15)
plt.show()
```
